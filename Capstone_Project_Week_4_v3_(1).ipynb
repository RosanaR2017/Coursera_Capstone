#### First, I cleaned the HTML headers by converting it to text and using remove_html tags function.


```python
def remove_html_tags(text):
    """Remove html tags from a string"""
    import re
    clean = re.compile(r'(<[^\>]*>)|(\n)|(\xa0)') 
    return re.sub(clean, '', text) 
```

Second, the header of the HTML are extracted:

```python
header=[]
for th in table.find_all('th'):
  header.append(str(th))
header=list(map(remove_html_tags,header[0:4]))
pprint("Obtained result: {}".format(header))
```

Third, the rows are also extracted:

```python
rows = []
for td in table.find_all('td'):
    rows.append(str(td).upper().replace(",",""))
rows=list(map(remove_html_tags,rows))
print("The row result is: {}".format(rows), sep=",")
```

From the Foursquare lab in the previous module, a get_category_type function from the Foursquare lab is used to extract the category of each venue.

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

Then, the JSON is processed, flattened and filtered in order to obatin ec_nearby_venues, which contains ecuador's nearby venues name, categories, latitude (lat) and longitude (lng).   

```python
ec_venues = ec_results['response']['groups'][0]['items']  #Processing JSON to extract the venues in Ecuador
ec_nearby_venues = pd.json_normalize(ec_venues) # flatten JSON
# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
ec_nearby_venues =ec_nearby_venues.loc[:, filtered_columns]

# filter the category for each row
ec_nearby_venues['venue.categories'] = ec_nearby_venues.apply(get_category_type, axis=1)

# clean columns
ec_nearby_venues.columns = [col.split(".")[-1] for col in ec_nearby_venues.columns]
ec_nearby_venues
```

Next, using function getNearbyVenues, the provinces in Ecuador are gotten within a radius of 100000.

```python
def getNearbyVenues(names, latitudes, longitudes, radius=100000):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        #print(name)      This part was commented in order to only view the dataframe.
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Provinces',   # This part of the function was changed to match our purpose.
                            'Latitude', 
                            'Longitude', 
                            'Venue', 
                            'Venue Latitude', 
                            'Venue Longitude', 
                            'Venue Category']
    
    return(nearby_venues)
```

```python
num_top_venues = 40

for venue in ecuador_grouped['Provinces']:
    print("----"+venue+"----")
    temp = ecuador_grouped[ecuador_grouped['Provinces'] == venue].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

Let's put all the 2x30 venues into a pandas dataframe.

First, let's write a function to sort the venues in descending order:

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

Now let's create the new dataframe and display the top 29 venues that are in each province.

```python
num_top_venues = 29

indicators = ['st', 'nd', 'rd']

# create columns according to number of province
columns = ['Provinces']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
provinces_venues_sorted = pd.DataFrame(columns=columns)
provinces_venues_sorted['Provinces'] = ecuador_grouped['Provinces']

for ind in np.arange(ecuador_grouped.shape[0]):
    provinces_venues_sorted.iloc[ind, 1:] = return_most_common_venues(ecuador_grouped.iloc[ind, :], num_top_venues)

provinces_venues_sorted
```

With this result, we notice that in Guayas the 1rst most common venue is a Golf course and for Pichincha is the Airport.
