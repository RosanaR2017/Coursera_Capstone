```python
import requests
```

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
source= requests.get(url).text
```

3. Importing pandas and reading the table

```python
import pandas as pd
```

```python
dfs = pd.read_html(source)
```

```python
df=dfs[0]
df
```

Filtering not assigned Borough

```python
df=df.query('Borough != "Not assigned"')
df.head()
```

```python
df=df.groupby("Postcode").agg(lambda x:','.join(set(x)))
df
```
